```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic

## Line Chart

```python
# data
x = np.array([x for x in range(1, 11)])
y = np.array(x**2)

# size
plt.figure(figsize=(5, 8), dpi=100)

# First line
plt.plot(x, y, label="Square", color="orange", linestyle="--", linewidth=2, marker=".", markersize=8, markeredgecolor="blue")

# Second line
z = np.array(x**3)
plt.plot(x, z, label="Cube", color="#6ecc73", linestyle="-", linewidth=2, marker=".", markersize=8, markeredgecolor="blue")

# Third line
plt.plot(x[:5], x[:5], label="Integer", color="lightblue", linestyle="-", linewidth=2, marker=".", markersize=8, markeredgecolor="blue")
plt.plot(x[4:], x[4:], label="", color="lightblue", linestyle="--", linewidth=2, marker=".", markersize=8, markeredgecolor="blue")

# description
plt.title("Line Chart", fontdict={"fontsize": 15})
plt.xlabel("Integer", fontdict={"fontsize": 10})
plt.ylabel("Polynom", fontdict={"fontsize": 10})

# number bar
plt.xticks(x)
# plt.yticks(y)

# legend
plt.legend()

# save
plt.savefig("line.png", dpi=300)

# show
plt.show()
```

## Bar Chart

```python
# data
labels = ["A", "B", "C"]
values = [1, 4, 2]

# First bars
bars = plt.bar(labels, values)

# pattern
patterns = ["/", "O", "*"]
for bar in bars:
  bar.set_hatch(patterns.pop(0))

# description
plt.title("Bar Chart", fontdict={"fontsize": 15})
plt.xlabel("Categories", fontdict={"fontsize": 10})
plt.ylabel("Value Counts", fontdict={"fontsize": 10})

# number bar
plt.yticks([y for y in range(1, 11)])

# show
plt.show()
```

# Real World Example

## Gas Prices

```python
df = pd.read_csv("/content/gas_prices.csv")
df
```

### Line chart

```python
for country in df:
  if country != "Year":
    plt.plot(df["Year"], df[country], label=country, linestyle="-", marker=".", markersize=5)

plt.xticks(df["Year"][::2].tolist() + [2010])
plt.yticks([y for y in range(0, 11)])

plt.title("Gas Prices (1990 - 2008)")
plt.xlabel("Year")
plt.ylabel("Price (US$)")

plt.legend()

plt.show()
```

## FIFA

```python
df = pd.read_csv("/content/fifa_data.csv")
df
```

### Histogram

```python
bins = [i for i in range(40, 101, 10)]

plt.hist(df["Overall"], bins=bins)

plt.xticks(bins)

plt.title("Distribution of Players Skills in FIFA 2018")
plt.xlabel("Skill Level")
plt.ylabel("Number of Players")

plt.show()
```

### Pie chart

```python
left = df.loc[df["Preferred Foot"] == "Left"].count()["ID"]
right = df.loc[df["Preferred Foot"] == "Right"].count()["ID"]

labels = ["Left", "Right"]

plt.pie([left, right], labels=labels, autopct="%.2f%%")

plt.title("Foot Preference of FIFA Players")

plt.show()
```

```python
df["Weight"] = [int(x.strip("lbs")) if type(x) == str else x for x in df["Weight"]]

light = df.loc[df["Weight"] < 125].count()["ID"]
light_medium = df.loc[(df["Weight"] >= 125) & (df["Weight"] < 150)].count()["ID"]
medium = df.loc[(df["Weight"] >= 150) & (df["Weight"] < 175)].count()["ID"]
medium_heavy = df.loc[(df["Weight"] >= 175) & (df["Weight"] < 200)].count()["ID"]
heavy = df.loc[df["Weight"] >= 200].count()["ID"]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ["Under 125", "125 - 149", "150 - 174", "175 - 199", "200 and Above"]
explode = (.4, 0, 0, 0, .4)

plt.style.use("ggplot")

plt.pie(weights, labels=labels, autopct="%.2f%%", pctdistance=0.8, explode=explode)

plt.title("Weight Distribution of FIFA Players (Pound)")

plt.show()
```

### Box Plot

```python
barcelona = df.loc[df["Club"] == "FC Barcelona"]["Overall"]
rm = df.loc[df["Club"] == "Real Madrid"]["Overall"]
ner = df.loc[df["Club"] == "New England Revolution"]["Overall"]

labels = ["FC Barcelona", "Real Madrid", "New England Revolution"]

boxes = plt.boxplot([barcelona, rm, ner], labels=labels, patch_artist=True, medianprops={"linewidth": 2})

for box in boxes["boxes"]:
  box.set(color="#4286f4", linewidth=2)

plt.title("Soccer Team Comparison")
plt.ylabel("FIFA Overall Rating")

plt.show()
```
