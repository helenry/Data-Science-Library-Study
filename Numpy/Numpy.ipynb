# Copying Array

### Problem

```python
a = np.array([1, 2, 3])
a
```

```python
b = a
b
```

```python
b[0] = 100
b
```

```python
a # a ikut berubah
```

### Solution

```python
a = np.array([1, 2, 3])
a
```

```python
b = a.copy()
b
```

```python
b[0] = 100
b
```

```python
a # a tidak berubah
```
